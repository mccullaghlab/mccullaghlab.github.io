# Thermodynamic Cycles: Solving exam problems and other examples

## Learning Goals

After this lesson, you should be able to:
1. Set up thermodynamic problems and solve them quickly in an exam setting.
2. Solve for pressure, volume, and temperature relationships on a thermodynamic cycle.
3. Solve for work, heat, change in internal energy, and change in enthalpy for a thermodynamic cycle process.
4. Compare extractable work from a process with an adiabat or isotherm.


## Tips for solving thermodynamic cycle exam problems
Follow these tips to solve thermodynamic cycle problems faster during an exam when time is limited.

1. Use or a make a table that lists each process and the initial and final values for pressure, temperature, and volume. Clearly identify initial and final values rather than using $V_1$ or $V_2$ to make plugging values into equations easier. 
2. Set up a blank table with each process listed and all the thermodynamic values you need to solve for.
3. Add any values you are given into the table.
4. Fill in any 'zero' values. For example, no work is extracted from an isochoric process and no heat from an adiabat. 
5. Set up a total row. State functions, such as $\Delta$ $U$ and $\Delta$ $H$, sum to zero in a cycle.
6. Use the relationship $w + q = \Delta U$. For an isotherm, $\Delta$ $U$ is zero so you can quickly find $w$ and $q$ by only solving for one and the other one must be equal in magnitude and opposite. 

Take advantage of the relationships in steps 4 and 5 to avoid plugging in and solving for each value. These relationships can also be used to check your work.

## Cycle A: Adiabatic, Isobaric, Isochoric
This 3 membered cycle begins with an adiabatic compression, then isobaric compression, and isochoric heating. We are given $n$ = 1, $V_1$ = 2.0 L, $V_2$ = 8.0 L, and $T_1$ = 300.0 K. To solve for $P_1$, $P_2$, and $T_2$ we will need to use adiabatic relationships and the ideal gas law. 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

#Define Variables
resolution = 1000
gamma = 5/3
n = 1
R=0.08206
T1 = 300
V1 = 2
V2 = 8
P_1=n*R*T1/V1
P_2 = P_1*(V1/V2)**gamma
T2 = (P_2*V2)/(n*R)
T3 = (P_2*V1)/(n*R)
V=np.linspace(V1,V2,resolution)
P_I = np.linspace(P_1,P_2,resolution)
fontsize = 16

#Adiabat
P_A = P_1*(V1/V)**gamma


#ISOBARIC
isobaric = np.linspace(P_2,P_2,resolution)

#ISOCHORIC
isochoric = np.linspace(V1,V1,resolution)

# Plot
fig, ax=plt.subplots(figsize=(8,6), dpi= 80, facecolor='w', edgecolor='k')
plt.text(V1, P_1, '($V_1$, $P_1$, $T_1$)', ha='left', va='top', fontsize=12)
plt.text(V1, P_2, '($V_1$, $P_2$, $T_3$)', ha='left', va='top', fontsize=12)
plt.text(V2, P_2, '($V_2$, $P_2$, $T_2$)', ha='right', va='top', fontsize=12)



# Curves with arrows
ax.plot(V, P_A, linewidth=4, label='Adiabatic')
arrow_A = patches.FancyArrowPatch((V[500], P_A[500]), (V[-499], P_A[-499]), color='black', mutation_scale=25)
ax.add_patch(arrow_A)

ax.plot(V, isobaric, linewidth=4, label='Isobaric')
arrow_B = patches.FancyArrowPatch((V[500], P_2), (V[499], P_2), color='black', mutation_scale=25)
ax.add_patch(arrow_B)

ax.plot(isochoric, P_I, linewidth=4, label='Isochoric')
arrow_C = patches.FancyArrowPatch((V1, P_I[500]), (V1, P_I[499]), color='black', mutation_scale=25)
ax.add_patch(arrow_C)

# Formatting
ax.set_xlabel(r"Volume ($\mathrm{L}$)", size=fontsize)
ax.set_ylabel("Pressure (atm)", size=fontsize)
ax.set_title("Cycle A", size=fontsize)
ax.grid(which='major', axis='both', color='#808080', linestyle='--')
ax.tick_params(axis='both', labelsize=fontsize)

# Common region
ax.fill_between(V, P_A, P_2, facecolor="green", alpha=0.5, interpolate=True)

# Legend
ax.legend(fontsize=12, loc='upper right')

plt.ylim(0, 15)
plt.show()
```

### Solving for P, V, T on Cycle A

To solve for $P_1$, use $T_1$, $V_1$, and the ideal gas law.
\begin{align}
P_1 &= \frac{nRT_1}{V_1} \\
   &= \frac{0.08206\cdot300}{2} \\
   &= 12.3 \text{ atm}
\end{align}

Use an adiabatic relationship to solve for $T_2$. At any point along the curve, $PV^{5/3}$ is constant. 
\begin{align}
P_1V_1^{5/3} &= P_2V_2^{5/3} \\
\Rightarrow P_2 &= P_1\cdot\left(\frac{V_1}{V_2}\right)^\frac{5}{3} \\
&= 12.3\cdot\left(\frac{2}{8}\right)^\frac{5}{3} \\
&= 1.2 \text{ atm}
\end{align}


To solve for $T_2$ use the ideal gas law and the values for $V_2$ and $P_2$.
\begin{align}
T_2 &= \frac{P_2V_2}{nR} \\
    &= \frac{1.2\cdot8}{0.08206} \\
    &= 118.9 \text{ K}
\end{align}

To solve for $T_3$ use the ideal gas law and the values for $V_1$ and $P_2$. 
\begin{align}
T_3 &= \frac{P_2V_1}{nR} \\
   &= \frac{1.2\cdot2}{0.08206} \\
   &= 29.7 \text{ K}
\end{align}

|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
| Adiabatic | 2.0 | 8.00 | 12.3 | 1.2 | 300.0 | 118.9 |
| Isobaric | 8.0 | 2.0 | 1.2 | 1.2 | 118.9 | 29.7 |
| Isochoric | 2.0 | 2.0 | 1.2 | 12.3 | 29.7 | 300.0 |

### Solving for work, heat, change in internal energy, and change in enthalpy on Cycle A

We will calculate values for work, heat, change in internal energy, and change in enthalpy using the following equations.


| Process  | $w$   |  $q$   | $\Delta$ U   | $\Delta$ H  |
| :----: | :--: | :----: | :---: | :-: |
| Adibatic    | $\frac{3}{2}nR(T_f-T_i)$ | $0$ | $\frac{3}{2}nR(T_f-T_i)$ | $\frac{5}{2}nR(T_f-T_i)$ |
| Isobaric  | $-(\frac{nRT_f}{V_f})\Delta V$ | $\Delta U - w$ | $\frac{3}{2}nR(T_f-T_i)$ | $\frac{5}{2}nR(T_f-T_i)$ |
| Isochoric  | $0$   | $\frac{3}{2}nR(T_i-T_f)$  | $\frac{3}{2}nR(T_f-T_i)$  | $\frac{5}{2}nR(T_f-T_i)$ |
| Total | $\frac{3}{2}nR(T_f-T_i)-(\frac{nRT_f}{V_f})\Delta V$ | $-\frac{3}{2}nR(T_f-T_i)+(\frac{nRT_f}{V_f})\Delta V$ |  $0$ | $0$ |

#### Adiabat
We want to solve for $w$, $q$, $\Delta$ $U$, and $\Delta$ $H$ from $(V_1, P_1, T_1)$ to $(V_2, P_2, T_2)$.
\begin{align}
q &= 0 \\
\\
w &= \frac{3}{2}nR(T_f-T_i) \\
&= \frac{3}{2}\cdot0.08206(118.9-300) \\
&= -22.3 \text{ L atm}\\
\\
\Delta U &= w \\
&= -22.3 \text{ L atm}\\
\\
\Delta H &= \frac{5}{2}nR(T_f-T_i) \\
&= \frac{5}{2}\cdot0.08206(118.9-300) \\
&= -37.2 \text{ L atm}\\
\end{align}

#### Isobaric
We want to solve for $w$, $q$, $\Delta$ $U$, and $\Delta$ $H$ from $(V_2, P_2, T_2)$ to $(V_1, P_2, T_3)$.
\begin{align}
w &= -(\frac{nRT_f}{V_f})\Delta V \\
&= -(\frac{0.08206\cdot29.7}{2})(2-8) \\
&=7.3 \text{ L atm}\\
\\
\Delta U &= \frac{3}{2}nR(T_f-T_i) \\
&= \frac{3}{2}\cdot0.08206(29.7-118.9) \\
&= -11.0 \text{ L atm}\\
\\
q &= \Delta U - w \\
&= -18.3 \text{ L atm}\\
\\
\Delta H &= \frac{5}{2}nR(T_f-T_i) \\
&= \frac{5}{2}\cdot0.08206(29.7-118.9) \\
&= -18.3 \text{ L atm}
\end{align}

#### Isochoric
We want to solve for $w$, $q$, $\Delta$ $U$, and $\Delta$ $H$ from $(V_1, P_2, T_2)$ to $(V_1, P_1, T_1)$.
\begin{align}
w &= 0 \\
\\
\Delta U &= \frac{3}{2}nR(T_f-T_i) \\
&= \frac{3}{2}\cdot0.08206(300-29.7) \\
&= 33.3 \text{ L atm}\\
\\
q &= \Delta U \\
&= 33.3 \text{ L atm}\\
\\
\Delta H &= \frac{5}{2}nR(T_f-T_i) \\
&= \frac{5}{2}nR(300-29.7) \\
&= 55.5 \text{ L atm}
\end{align}

We want to solve for $w$, $q$, $\Delta$ $U$, and $\Delta$ $H$ from $(V_1, P_1, T_1)$ to $(V_2, P_2, T_1)$.

Along an isotherm, temperature is constant which is why $\Delta$ $U$ and $\Delta$ $H$ are 0.
\begin{align}
  \Delta U &= 0 \\
  \Delta H &= 0 \\
\end{align}
To calulate work and heat use the equations that have been derived in previous notebooks.
\begin{align}
        w &= nRT_2\ln(\frac{V_i}{V_f}) \\
        &= 0.08206\cdot300.0\ln(\frac{2}{8}) \\
        &= - 34.1 \text{ L atm}\\
        \\
        q &= nRT_2\ln(\frac{V_f}{V_i}) \\
        &= 0.08206\cdot300.0\ln(\frac{8}{2}) \\
        &= 34.1 \text{ L atm}
\end{align}
To solve for heat, we could have also recognized that the sum of work and heat must equal zero since $w + q = \Delta U$


To compare the extracted work, we need to set up an equation that relates work to difference in volume and the starting point. In this example, each cycle begins at the same starting point and has the same change in volume. We will derive work equations for each process in terms of $V_i$ and $V_f$. 

#### Work for Cycle A
The work from Cycle A comes from the adiabat ($w_A$) and the isobaric process ($w_I$). There is no work extracted from the isochoric process. We will manipulate the work equation for each separately then add them together to have an equation that represents work for Cycle A. The equation we used to solve for work earlier on an adiabat is:
\begin{align}
w_A &= \frac{3}{2}nR(T_f-T_i)\\
\end{align}

We will use the adiabat relationship $TV^\frac{2}{3}$ to relate work to volume, not the temperature.
\begin{align}
T_iV_i^\frac{2}{3} &= T_fV_f^\frac{2}{3} \\
\Rightarrow T_f &= T_i\cdot\left(\frac{V_i}{V_f}\right)^\frac{2}{3} \\
\end{align}

Substract $T_i$ from each side substitutue into the work equation.
\begin{align}
T_f -T_i &= T_i\cdot\left(\frac{V_i}{V_f}\right)^\frac{2}{3} -T_i \\
\Rightarrow w_A &= \frac{3}{2}nRT_i\cdot\left(\frac{V_i}{V_f}\right)^\frac{2}{3} -T_i
\end{align}
We also need to work extracted from the isobaric process. Work from an isobaric process is calculated from $w = -P\Delta V$. We will need to define $\Delta V$ in terms of the whole cycle and define $P$. 
For an adiabat $P$ will be determined by the relationship $PV^{5/3}$
\begin{align}
P_iV_i^{5/3} &= P_fV_f^{5/3} \\
\Rightarrow P_f &= P_i\cdot\left(\frac{V_i}{V_f}\right)^\frac{5}{3} \\
\end{align}
Initial pressure is determined by the starting point. Since we are given starting point values for $V$ and $T$ we can solve for $P$ using the ideal gas law and substitute this relationship.
\begin{align}
P_i &= \frac{nRT}{V_i} \\
\Rightarrow w_I &= -\frac{nRT}{V_i}\cdot\left(\frac{V_i}{V_f}\right)^\frac{5}{3}\cdot\Delta V \\
\end{align}
Finally, $\Delta V$ refers to the change in volume in the isobaric process. In the three-membered cycle used in this example, the final volume of the isobaric process is equal to the initial volume. We can substitute $\Delta V = V_i - V_f$.
The resulting equation for work from the isobaric process of Cycle A is 
\begin{align}
w_I &= -\frac{nRT}{V_i}\cdot\left(\frac{V_i}{V_f}\right)^\frac{5}{3}\cdot(V_i - V_f) \\
\end{align}
To get the total work from Cycle A we will simplify and sum $w_A$ and $w_I$
\begin{align}
w &= nRT_i\cdot\left[\frac{3}{2}\left(\frac{V_i}{V_f}\right)^\frac{2}{3}-\frac{3}{2} -\frac{V_i - V_f}{V_i}\cdot\left(\frac{V_i}{V_f}\right)^\frac{5}{3}\right] \\
&= nRT_i\cdot\left[\frac{5}{2}\left(\frac{V_i}{V_f}\right)^\frac{2}{3}-\frac{3}{2} - \left(\frac{V_i}{V_f}\right)^\frac{5}{3}\right]
\end{align}

To prove this equation works, below is the calculation for Cycle A that we solved earlier. The values match.

```python
# compute work for process A in units of nR
def wA(Ti,Vi_Vf):
    return Ti*(2.5*(Vi_Vf)**(2/3) - 1.5 - (Vi_Vf)**(5/3))

print(wA(300,2/8)*0.08206)
```

#### Work for Cycle B
We will do the same for Cycle B. The equation we used for work on an isotherm, $w = nRT\ln(\frac{V_i}{V_f})$. This relationship is already in terms of initial and final volume. We also need to rearrange the isobaric process like we did for Cycle A. Work from an isobaric process is calculated from $w = -P\Delta V$. We can use the ideal gas law to substitute for $P$.
\begin{align}
P_f &= \frac{nRT}{V_f} \\
\Rightarrow w &= \frac{nRT}{V_f}\cdot(V_i - V_f)
\end{align}
Just as before we will sum the two processes to get total extractable work for Cycle B.
\begin{align}
w &= nRT\left[ln\left(\frac{V_i}{V_f}\right)-\frac{V_i-V_f}{V_f}\right] \\
&= nRT\left[ln\left(\frac{V_i}{V_f}\right)+1 - \frac{V_i}{V_f}\right]
\end{align}

To prove this equation is valid, below is the calculation for Cycle B that we solved earlier. 


```python
# compute work for process A in units of nR
import numpy as np
def wB(Ti,Vi_Vf):
    return Ti*(np.log(Vi_Vf) + 1 - Vi_Vf)

print(wB(300,2/8)*0.08206)
```

#### Ratio of extracted work from Cycle A or B based on change in volume
Now that we have derived equations for work for each of these processes that are dependent on change in volume and starting at the same value, we can compare the work extracted from each process based on the ratio of volumes. The plot below shows that the difference in work extracted from Cycle A and Cycle B depends on the ratio between final and initial volume. If the ratio of final to initial volume is less than about 3.5, then Cycle B extracts more work than Cycle A. 


```python
import matplotlib.pyplot as plt
%matplotlib inline
# Plot
fontsize=16
fig, ax=plt.subplots(figsize=(8,6), dpi= 80, facecolor='w', edgecolor='k')
Ti = 300
Vi_Vf = np.arange(1,0.1,-0.01)
ax.plot(1/Vi_Vf, (wA(Ti,Vi_Vf) - wB(Ti,Vi_Vf))*0.08206, linewidth=4, label=r'w$_B$-w$_A$')
# Formatting
ax.set_xlabel(r'V$_f$/V$_i$', size=fontsize)
ax.set_ylabel(r'w$_A$-w$_B$ (nR)', size=fontsize)
ax.grid (which='major', axis='both', color='#808080', linestyle='--')
plt.tick_params(axis='both',labelsize=fontsize)
```
